```python
training_dataset = pd.read_csv(datasets['train.csv'])
test_dataset = pd.read_csv(datasets['test.csv'])
```

```python
# Clean the training dataset
display(training_dataset.shape)
training_dataset.dropna(inplace = True)

x = training_dataset['x']
y = training_dataset['y']
x_sum = x.sum()
y_sum = y.sum()
n = training_dataset.shape[0]
x_square_sum = (x**2).sum()
display(training_dataset.shape)
```

```python
# Calculate the slope and intercept
b1 = ((x*y).sum() - (x_sum*y_sum/n))/(x_square_sum - (x_sum**2/n))
b0 = (y_sum/n) - (b1 * (x_sum/n))
print('b0 = ', b0,' b1 = ', b1)
```

```python
def predict(x):
    return b0 + (b1*x)
```

```python
y_predicted = predict(test_dataset['x'])
```

```python
plt.scatter(test_dataset['x'], y_predicted, color = 'black') # Plot the predicted dataset datapoints
plt.plot(test_dataset['x'], y_predicted) # Plot the line that was fitted
```
